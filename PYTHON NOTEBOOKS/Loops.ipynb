# for loop 

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

```python
for i in range(3):
    print(i)  # Output: 0, 1, 2
```

# while loop 

```python
count = 0
while count < 3:
    print(count)
    count += 1
```

```python
for i in range(5):
    if i == 3:
        break  # Stops the loop at 3
    print(i)
```

# break, continue and pass

```python
l = [1,2,3,4,5]

for item in l:
    pass

print("Hell0")
```

```python
l = ["apples", "bananas", "tomatoes"]
for item in l:
    if ( item == "bananas"):
        break # stops the loop right here
    print(item)
```

```python
l = ["apples", "bananas", "tomatoes"]
for item in l:
    if ( item == "bananas"):
        continue # skips this particular iteration of the loop
    print(item)
```
